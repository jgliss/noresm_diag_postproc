# Interface for downloading ASCII files as CSV from online resource

```python
from helper_funcs import download_cams_diag_result_csv, read_and_merge_all
import os

SAVE_DIR = "./output/download/"

#list of table URLs that are supposed to be downloaded (here only one)
URLS = ['http://ns2345k.web.sigma2.no/noresm_diagnostics/NF1850_CTRL_f19_f19_r265_20180412/CAM_DIAG/yrs5to30-obs/set1/table_GLBL_ANN_obs.asc',
       ]
if not os.path.exists(SAVE_DIR):
    os.mkdir(SAVE_DIR)
```

## Download csv's into local directory

Now download all tables and save in specified output directory

```python
downloaded = []
for URL in URLS:
    try:
        downloaded.append(download_cams_diag_result_csv(URL, 
                                                        save_dir=SAVE_DIR))
    except Exception as e:
        print("Failed to download table. Error message: {}".format(repr(e)))     
```

## Check if everything worked (read csv's and create dataframe)

Now try to read the downloaded files

```python
table = read_and_merge_all(downloaded)
table
```
